```python
import re

def extract_strings(filename, min_length=4):
    with open(filename, "rb") as f:
        data = f.read()
    strings = re.findall(rb"[A-Za-z0-9_./\\-]{%d,}" % min_length, data)
    strings_list = []
    for s in strings:
        strings_list.append(s.decode(errors="ignore"))

    sorted_strings = sorted(set(strings_list))
    for s in sorted_strings:
        print(s)

if __name__ == "__main__":
    extract_strings("nn_model_secure_dill.pkl")
```

```python
def find_context(filename, keywords, context=30):
    with open(filename, "rb") as f:
        data = f.read()
    for word in keywords:
        word_bytes = word.encode()
        idx = 0
        while True:
            idx = data.find(word_bytes, idx)
            if idx == -1:
                break
            start = max(0, idx - context)
            end = min(len(data), idx + len(word_bytes) + context)
            snippet = data[start:end]
            print(f"Palavra '{word}' encontrada na posio {idx}:")
            print(snippet.decode(errors="replace"))
            print("-" * 40)
            idx += len(word_bytes)

find_context("nn_model_secure_dill.pkl", ["sleep", "sleeping", "asleep"], context=40)
```

```python
import gzip
import dill as pickle

# Descompactar o arquivo .gz
with gzip.open('model.pkl.gz', 'rb') as f_in:
    data = pickle.load(f_in)

# Verificar o tipo do objeto carregado
print("Tipo do objeto:", type(data))

# Se for dict, listar as chaves
if isinstance(data, dict):
    print("Entradas do sistema de arquivos:")
    for key in data.keys():
        print(key)
elif isinstance(data, list):
    print("Itens do sistema de arquivos:")
    for i, item in enumerate(data):
        print(f"Item {i}: {type(item)}")
else:
    print("Objeto no reconhecido como dict ou list. Explore os atributos:")
    print(dir(data))
```
